```python
class DataCleaner:
    def __init__(self, file_path):
        self.df = pd.read_csv(r'C:\Users\Arindham Krishna\Desktop\Data Engineer MIT\Data Sets\NYC\AB_NYC_2019.csv')

    def drop_missing_values(self, columns):
        """Remove rows where specified columns have missing values."""
        self.df.dropna(subset=columns, inplace=True)
        return self

    def remove_duplicates(self):
        """Drop duplicate listings based on 'id'."""
        self.df.drop_duplicates(subset=['id'], inplace=True)
        return self
    
    def get_cleaned_df(self):
        return self.df
    
cleaner = DataCleaner("AB_NYC_2019.csv")
cleaned_df = (
    cleaner.drop_missing_values(['reviews_per_month'])
           .remove_duplicates()
           .get_cleaned_df()
)

cleaned_df.head()
```

```python
import requests

class NewsAPI:
    def __init__(self):
        """Initialize with API base URL and API key"""
        self.base_url = "https://newsapi.org/v2/everything"
        self.api_key = "5869a08a34cc47b6b34af2d39d63eaee"  
        
    def fetch_data(self, keyword):
        """Fetch news data based on a keyword"""
        params = {
            "q": keyword,  # User keyword
            "apiKey": self.api_key  # API Key added in parameters
        }

        response = requests.get(self.base_url, params=params)

        if response.status_code == 200:
            raw_data = response.json()  
            return self.normalized_data(raw_data)
        else:
            print(f"Error! Check Status Code: {response.status_code}")
            return None
        
    def normalized_data(self,raw_data):
        self.df = pd.json_normalize(raw_data, 'articles', sep = '_')
        return self.df


news_extractor = NewsAPI()
data = news_extractor.fetch_data("Trump")  

data
```
